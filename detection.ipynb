```python
from ultralytics import YOLO
```

```python
# Load the model.
model = YOLO('yolov8n.pt')
```

```python
# Training.
results = model.train(
   data='data.yaml',
   imgsz=1280,
   epochs=50,
   batch=8,
   name='yolov8n_shai_50e')

# yolo task=detect mode=train model=yolov8n.pt imgsz=1280 data=pothole_v8.yaml epochs=50 batch=8 name=yolov8n_v8_50e
```

```python
model_trained = YOLO("runs/detect/yolov8n_shai_50e/weights/best.pt")
```

```python
from PIL import Image
import PIL
  
images_list = ['1.jpg', '2.jpg', '3.jpg']
               
for im in images_list:
    image = Image.open(im)
    image.thumbnail((640, 640))
    image = image.rotate(270, PIL.Image.NEAREST, expand = 1)
    image = image.save(im+'resized.jpg')
    # image.show()
```

```python
images_list = ['1.jpg', '2.jpg', '3.jpg']

for im in images_list:
    results = model_trained.predict(source=im, line_thickness=3, conf=0.1, save=True)
```

```python
images_list = ['1.jpgresized.jpg', '2.jpgresized.jpg', '3.jpgresized.jpg']

for im in images_list:
    results = model_trained.predict(source=im, line_thickness=1, conf=0.1, save=True)
```
